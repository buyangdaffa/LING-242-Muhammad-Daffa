### Creutz and Lagus (2002): unsupervised discovery of morphemes

```python
# Downloads the Morfessor tool version 2.0.6 in a compressed archive format.
!wget http://morpho.aalto.fi/projects/morpho/morfessor-2.0.6.tar.gz
```

```python
# Extracts the contents of the Morfessor tool archive.
!tar xvfz morfessor-2.0.6.tar.gz
```

```python
# Changes the directory to the extracted Morfessor tool directory.
%cd morfessor-2.0.6
```

```python
# Installs Morfessor tool by executing its setup script.
!python3 setup.py install
```

```python
# Downloads the preprocessed Finnish training data in a compressed format.
!wget https://github.com/jungyeul/computational-tools-for-linguistic-analysis-ubc/raw/main/labs/lab6/fin.train.lowered.tok.gz
```

```python
# Decompresses the downloaded Finnish training data.
!gzip -d fin.train.lowered.tok.gz
```

```python
# Selects the first 1000 lines from the decompressed Finnish training data and saves it as "fin.train.lowered.tok.1K".
!head -n 1000 fin.train.lowered.tok > fin.train.lowered.tok.1K
```

```python
# Trains Morfessor model using the selected subset of Finnish training data and saves the model as "model_1K.bin".
!./scripts/morfessor-train fin.train.lowered.tok.1K -s model_1K.bin
```

```python
# Downloads the Finnish test data.
!wget https://github.com/jungyeul/computational-tools-for-linguistic-analysis-ubc/raw/main/labs/lab6/fin.test
```

```python
# Segments the Finnish test data using the trained Morfessor model and saves the segmented output as "fin.test.result".
!./scripts/morfessor-segment -l model_1K.bin fin.test > fin.test.result
```

```python
# Downloads the gold standard test data for evaluating Finnish segmentation.
!wget https://github.com/jungyeul/computational-tools-for-linguistic-analysis-ubc/raw/main/labs/lab6/fin.test_gold
```

```python
# Combines the Finnish test data, segmented output, and gold standard data side by side for comparison.
!paste fin.test fin.test.result fin.test_gold
```
